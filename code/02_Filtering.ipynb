```python
import gwosc.datasets as g
from gwpy.timeseries import TimeSeries
from gwpy.plot import Plot
from gwpy.signal import filter_design

event = 'GW150914-v1'
gps = g.event_gps(event)
dat_1 = TimeSeries.fetch_open_data("H1", gps-5, gps+5)
dat_2 = TimeSeries.fetch_open_data("L1", gps-5,gps+5)


white_1 = dat_1.whiten(fftlength=1/16)
white_2 = dat_2.whiten(fftlength=1/16)
band = filter_design.bandpass(35, 350, white_1.sample_rate)     #frequencies from Ann. Phys. (Berlin) 529, No. 1–2, 1600209 (2017) / DOI 10.1002/andp.201600209
not_ = [filter_design.notch(f, white_1.sample_rate) for f in [60,120,180]]
com = filter_design.concatenate_zpks(band, *not_)

dat_1f = white_1.filter(com,filtfilt=True)
dat_2f = white_2.filter(com,filtfilt=True)
dat_2f.shift('6.9ms')
dat_2f *= -1

plot = Plot(figsize=[12, 4])
ax = plot.gca()
ax.plot(dat_1f, label="H1", color="gwpy:ligo-hanford")
ax.plot(dat_2f, label="L1", color="gwpy:ligo-livingston")
ax.set_title("LIGO strain data GW150914")
ax.set_xlim(gps-.1,gps+.1)
ax.set_xscale('seconds',epoch=gps)
ax.set_ylabel("strain amplitude")
ax.legend()
plot.show()
```
